```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from tslearn.clustering import TimeSeriesKMeans
from kneed import KneeLocator
from sklearn.preprocessing import MaxAbsScaler
from statsmodels.tsa.statespace.sarimax import SARIMAX
from sklearn.metrics import mean_squared_error, mean_absolute_percentage_error, silhouette_score
import itertools
from datetime import datetime

# Load the dataset
data = pd.read_csv(r"D:\Data Science & AI\Projects\Main Projects\Medical Inventery Management\Medical Inventory Optimaization Dataset.csv")

# Define a function to parse dates with different formats
def parse_date(date_str):
    for fmt in ("%m/%d/%Y", "%d-%m-%Y"):
        try:
            return datetime.strptime(date_str, fmt)
        except ValueError:
            continue
    return pd.NaT

# Apply date parsing function
data['Dateofbill'] = data['Dateofbill'].apply(parse_date)

# Convert 'Dateofbill' from string to datetime format
data['Dateofbill'] = pd.to_datetime(data['Dateofbill'].dt.strftime('%Y-%m-%d'))

# Type casting
data["Patient_ID"] = data["Patient_ID"].astype('str')
data["Final_Sales"] = data["Final_Sales"].astype('float32')
data["Final_Cost"] = data["Final_Cost"].astype('float32')

# Handling Duplicates
data = data.drop_duplicates()

# Handling missing values
data.replace('', pd.NA, inplace=True)
group_cols = ['Typeofsales', 'Specialisation', 'Dept']

# Impute missing values in columns based on the mode of the group
for col in ['Formulation', 'DrugName', 'SubCat', 'SubCat1']:
    data[col] = data.groupby(group_cols)[col].transform(
        lambda x: x.fillna(x.mode().iloc[0]) if not x.mode().empty else x)

# Drop remaining missing values
data.dropna(inplace=True)
data = data.reset_index(drop=True)

# Remove rows with return type of transaction
data = data[data['Typeofsales'] != 'Return']

# Define a function to map dates to monthly periods
def get_monthly_period(date):
    return date.to_period('M').start_time

# Apply the function to create a monthly period column
data['Monthly_Period'] = data['Dateofbill'].apply(get_monthly_period)

# Aggregate sales data monthly
monthly_data = data.groupby(['DrugName', 'Monthly_Period'])['Quantity'].sum().reset_index()

# Find top 20 drugs based on total quantity sold
top_20_drugs = data.groupby('DrugName')['Quantity'].sum().nlargest(20).reset_index()

# Filter data to include only the top 20 drugs
top_20_drug_names = top_20_drugs['DrugName'].tolist()
filtered_data = monthly_data[monthly_data['DrugName'].isin(top_20_drug_names)]

# Pivot the table for clustering
pivoted_data = filtered_data.pivot(index='Monthly_Period', columns='DrugName', values='Quantity').fillna(0)

# Create individual plots for each drug
for drug in top_20_drugs['DrugName']:
    plt.figure(figsize=(10, 6))
    plt.plot(pivoted_data.index, pivoted_data[drug], marker='o')
    plt.xlabel('Date')
    plt.ylabel('Quantity Sold')
    plt.title(f'Monthly Sales for {drug}')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.show()

# Prepare data for clustering
X = pivoted_data.T

# Scale the data
scaler = MaxAbsScaler()
X_scaled = scaler.fit_transform(X)

# Determine the optimal number of clusters
inertia = []
silhouette_scores = []

for k in range(2, 11):
    kmeans = TimeSeriesKMeans(n_clusters=k, metric="dtw", verbose=True, random_state=0)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_, metric="euclidean"))

kl = KneeLocator(range(2, 11), inertia, curve="convex", direction="decreasing")
optimal_clusters = kl.knee

print(f'Optimal number of clusters: {optimal_clusters}')

# Perform clustering
kmeans = TimeSeriesKMeans(n_clusters=optimal_clusters, metric="dtw")
clusters = kmeans.fit_predict(X_scaled)

# Create a DataFrame to hold the cluster assignments for each drug
drug_clusters = pd.DataFrame({'DrugName': pivoted_data.columns, 'Cluster': clusters})

# Plot clusters
for cluster in range(optimal_clusters):
    plt.figure(figsize=(10, 6))
    cluster_drugs = drug_clusters[drug_clusters['Cluster'] == cluster]['DrugName']
    for drug in cluster_drugs:
        plt.plot(pivoted_data.index, pivoted_data[drug], label=drug, marker='o')
    plt.xlabel('Date')
    plt.ylabel('Quantity Sold')
    plt.title(f'Cluster {cluster} Sales')
    plt.xticks(rotation=45)
    plt.legend()
    plt.grid(True)
    plt.show()

# Calculate silhouette score for the chosen number of clusters
silhouette_avg = silhouette_score(X_scaled, clusters, metric="euclidean")
print(f'Silhouette Score for {optimal_clusters} clusters: {silhouette_avg}')

# Forecast for each cluster
for cluster in range(optimal_clusters):
    cluster_drugs = drug_clusters[drug_clusters['Cluster'] == cluster]['DrugName']

    # Use the first drug in the cluster for forecasting
    top_drug = cluster_drugs.iloc[0]
    drug_data = pivoted_data[top_drug].reset_index()
    drug_data.columns = ['Monthly_Period', 'Quantity']
    
    # Set 'Monthly_Period' as the index
    drug_data.set_index('Monthly_Period', inplace=True)

    # Split the data into training and testing sets
    train_size = int(len(drug_data) * 0.8)
    train_data, test_data = drug_data.iloc[:train_size], drug_data.iloc[train_size:]

    # Find the best SARIMA parameters
    best_mape = float("inf")
    best_pdq = None
    best_seasonal_pdq = None

    p = d = q = range(0, 3)
    pdq = list(itertools.product(p, d, q))
    seasonal_pdq = [(x[0], x[1], x[2], 12) for x in list(itertools.product(p, d, q))]

    for param in pdq:
        for param_seasonal in seasonal_pdq:
            try:
                mod = SARIMAX(train_data['Quantity'],
                              order=param,
                              seasonal_order=param_seasonal,
                              enforce_stationarity=False,
                              enforce_invertibility=False)
                results = mod.fit(disp=False)
                forecast = results.get_forecast(steps=len(test_data))
                predicted_mean = forecast.predicted_mean
                mape = mean_absolute_percentage_error(test_data['Quantity'], predicted_mean)
                if mape < best_mape:
                    best_mape = mape
                    best_pdq = param
                    best_seasonal_pdq = param_seasonal
            except:
                continue

    print(f'Best SARIMA parameters for Cluster {cluster}: {best_pdq} x {best_seasonal_pdq}')
    print(f'Best MAPE for Cluster {cluster}: {best_mape}')

    # Fit the best SARIMA model
    best_sarima_model = SARIMAX(train_data['Quantity'],
                                order=best_pdq,
                                seasonal_order=best_seasonal_pdq,
                                enforce_stationarity=False,
                                enforce_invertibility=False)
    best_sarima_fit = best_sarima_model.fit(disp=False)

    # Forecast the future sales
    n_forecast = len(test_data)
    forecast = best_sarima_fit.get_forecast(steps=n_forecast)
    forecast_index = test_data.index

    # Plot the forecast
    plt.figure(figsize=(10, 6))
    plt.plot(train_data.index, train_data['Quantity'], label='Train')
    plt.plot(test_data.index, test_data['Quantity'], label='Test')
    plt.plot(forecast_index, forecast.predicted_mean, label='Forecast')
    plt.fill_between(forecast_index, forecast.conf_int()['lower Quantity'], forecast.conf_int()['upper Quantity'], color='k', alpha=0.1)
    plt.xlabel('Date')
    plt.ylabel('Quantity Sold')
    plt.title(f'SARIMA Forecast for Cluster {cluster} ({top_drug})')
    plt.legend()
    plt.grid(True)
    plt.show()

    # Evaluate the model
    mse = mean_squared_error(test_data['Quantity'], forecast.predicted_mean)
    mape = mean_absolute_percentage_error(test_data['Quantity'], forecast.predicted_mean)
print(f'Mean Squared Error for Cluster {cluster}: {mse}')
print(f'Mean Absolute Percentage Error for Cluster {cluster}: {mape}')
print()
```
