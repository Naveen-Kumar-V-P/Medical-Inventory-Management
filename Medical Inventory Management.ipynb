```python
# Assuming your DataFrame is named 'medical' and 'Final_Sales' is a numerical column
numeric_column = 'Final_Sales'

# Calculate quartiles
q1 = data[numeric_column].quantile(0.25)
q3 = data[numeric_column].quantile(0.75)

# Calculate IQR
iqr = q3 - q1

# Identify outliers (below Q1-1.5*IQR or above Q3+1.5*IQR)
lower_bound = q1 - 1.5 * iqr
upper_bound = q3 + 1.5 * iqr

outliers_indices = np.where((data[numeric_column] < lower_bound) | (data[numeric_column] > upper_bound))[0]

# Access outlier data points (optional)
outlier_data = data.iloc[outliers_indices]

# Print or analyze outlier data (optional)
print(outlier_data)
```

```python
data = pd.read_csv(r"D:\Data Science & AI\Projects\Main Projects\Medical Inventery Management\Articles\data.csv")
```

```python

drug_sales = data.groupby('DrugName')['Quantity'].sum().sort_values(ascending=False)

top_selling_drug = drug_sales.index[0]
if 'Final_Cost' in data.columns:
  
    top_selling_drug_cost = data[data['DrugName'] == top_selling_drug]['Final_Cost'].iloc[0]
else:
    top_selling_drug_cost = None 

print("Most Sold Drug Name:", top_selling_drug)
if top_selling_drug_cost is not None:
    print("Cost per Unit:", top_selling_drug_cost)
```

```python
data['Profit'] = (data['Quantity'] * (data['Final_Sales'] - data['Final_Cost']))

drug_profits = data.groupby('DrugName')['Profit'].sum().sort_values(ascending=False)

top_profitable_drug = drug_profits.index[0]

print("Most Profitable Drug:", top_profitable_drug)
```
